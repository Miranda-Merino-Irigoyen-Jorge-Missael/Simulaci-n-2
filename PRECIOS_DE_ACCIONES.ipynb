\section{Definición de Rendimiento o Retorno en el Cálculo de Precios de Acciones}

El rendimiento o retorno de una acción representa la variación porcentual en el valor de la inversión durante un período determinado.  Se puede calcular de diversas maneras, dependiendo del horizonte temporal y los factores considerados.  A continuación se presentan algunas definiciones comunes:

\subsection{Retorno Simple}

El retorno simple se calcula como la diferencia entre el precio final y el precio inicial de la acción, dividido por el precio inicial.

\begin{equation}
\text{Retorno Simple} = \frac{P_t - P_0}{P_0}
\end{equation}

Donde:

* $P_t$: Precio de la acción al final del período.
* $P_0$: Precio de la acción al inicio del período.

Este retorno se expresa generalmente como un porcentaje.  Es importante destacar que el retorno simple no considera los dividendos que pueda haber pagado la acción durante el período.


\subsection{Retorno Total o Retorno Logarítmico}

El retorno logarítmico, también conocido como retorno continuo o compuesto, se calcula como el logaritmo natural del ratio entre el precio final y el precio inicial.

\begin{equation}
\text{Retorno Logarítmico} = \ln\left(\frac{P_t}{P_0}\right)
\end{equation}

Este tipo de retorno es útil para analizar períodos cortos de tiempo y facilita el cálculo del rendimiento acumulado a lo largo de varios períodos. La ventaja del retorno logarítmico es que permite sumar los rendimientos de diferentes periodos para obtener el rendimiento total, a diferencia del retorno aritmético.

\subsection{Retorno Total (Incluyendo Dividendos)}

Para una visión más completa del rendimiento, se debe considerar el pago de dividendos.  El retorno total incluye tanto la variación en el precio de la acción como los dividendos recibidos.

\begin{equation}
\text{Retorno Total} = \frac{P_t + D - P_0}{P_0}
\end{equation}

Donde:

* $D$: Dividendos recibidos durante el período.


\subsection{Consideraciones Adicionales}

* **Horizonte temporal:** El período de tiempo considerado influye en la interpretación del rendimiento. Un retorno alto en un período corto no necesariamente implica un rendimiento alto a largo plazo.

* **Riesgo:** El rendimiento de una acción está ligado al riesgo.  Acciones con mayor potencial de rendimiento suelen ser más volátiles y conllevan un mayor riesgo.

* **Inflación:** Es importante considerar la inflación al evaluar el rendimiento real de una inversión. El rendimiento nominal debe ajustarse por inflación para obtener el rendimiento real.

```python
# prompt: Graficame los precios de las tres empresas para 5 años, tambien graficame sus retornos y su retorno con logaritmo de manera de histograma

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Acciones a graficar
tickers = ["TSLA", "META", "UBER"]
start_date = "2018-12-13"  # Changed start date for 5 years of data

# Function to fetch and plot stock prices
def plot_stock_prices(tickers, start_date):
    plt.figure(figsize=(12, 6))
    for ticker in tickers:
        try:
            stock = yf.Ticker(ticker)
            df = stock.history(start=start_date, interval="1d")
            if df.empty:
                print(f"No se encontraron datos para {ticker} desde {start_date}.")
                continue

            close_prices = df["Close"]
            plt.plot(close_prices.index, close_prices.values, label=ticker)

        except Exception as e:
            print(f"Ocurrió un error al obtener datos de {ticker}: {e}")

    plt.xlabel("Fecha")
    plt.ylabel("Precio de Cierre")
    plt.title("Precios de Cierre de Acciones")
    plt.grid(True)
    plt.legend()
    plt.show()

plot_stock_prices(tickers, start_date)


# Function to calculate and plot simple returns and log returns as histograms
def plot_returns(tickers, start_date):
    for ticker in tickers:
        try:
            stock = yf.Ticker(ticker)
            df = stock.history(start=start_date, interval="1d")
            if df.empty:
                print(f"No se encontraron datos para {ticker} desde {start_date}.")
                continue

            close_prices = df["Close"]
            simple_returns = close_prices.pct_change().dropna()
            log_returns = np.log(close_prices) - np.log(close_prices.shift(1))
            log_returns = log_returns.dropna()

            plt.figure(figsize=(12, 4))
            plt.subplot(1, 2, 1)  # Two subplots in one row
            plt.hist(simple_returns, bins=30, alpha=0.5, label='Simple Returns')
            plt.title(f"Histograma de Retornos Simples para {ticker}")
            plt.xlabel("Retorno Simple")
            plt.ylabel("Frecuencia")

            plt.subplot(1, 2, 2)
            plt.hist(log_returns, bins=30, alpha=0.5, label='Log Returns', color='orange')
            plt.title(f"Histograma de Retornos Logarítmicos para {ticker}")
            plt.xlabel("Retorno Logarítmico")
            plt.ylabel("Frecuencia")

            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"Ocurrió un error al procesar datos de {ticker}: {e}")

plot_returns(tickers, start_date)
```
